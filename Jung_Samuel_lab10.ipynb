```python
#button countdown
#code from https://python-course.eu/tkinter/buttons-in-tkinter.php

counter = 0 
def counter_label(label):
    counter = 0
    def count():
        global counter #global variable declaration 
        counter += 1
        label.config(text=str(counter))
        label.after(1000, count) #after(parent, ms, function = None, *args)
    count()
 
 
root = Tk()
root.title("Counting Seconds")
label = Label(root, fg="dark green")
label.pack()
counter_label(label)
button = Button(root, text='Stop', width=25, command=root.destroy) #destroy(parent, command, widgetobject.destroy) 
button.pack()
root.mainloop()
```

```python
#basic dropdown menu
from tkinter import *

root = Tk()
root.title("Dropdown practice")
root.geometry("400x400")

item_list = ["Yellow", "Blue", "Green", "Purple", "Violet", "Black"]

def chosen_option():
    color = value.get()
    result_label = Label(root, text=f"You've chosen {color}")
    root.configure(background=color)
    result_label.pack()
    
value = StringVar()
value.set("Select a Color")
    
dropdown = OptionMenu(root, value, *item_list)
choose_button = Button(root, text="Click to choose", command=chosen_option)

dropdown.pack()
choose_button.pack()

root.mainloop()
```

```python
#radio buttons & Tkinter variables

root = Tk()
root.title("Radio Buttons")

#intvar helps it keep track of the variable at any one moment
r = IntVar()
r.set("2") #sets it to the second value

def clicked(value): 
    mylabel = Label(root, text=value)
    mylabel.pack()

Radiobutton(root, text="Choice1", variable=r, value=1, command=lambda: clicked(r.get())).pack()
Radiobutton(root, text="Choice2", variable=r, value=2, command=lambda: clicked(r.get())).pack()

root.mainloop()
```

```python
#GUI 1
import requests
import random
from tkinter import *

root = Tk()

# Create a function to get random list of people
def get_people():
    response = requests.get('https://api.quotable.io/authors?limit=50')
    # Extracted the names of the people from the JSON response
    people = [person['name'] for person in response.json()['results']]
    # random.shuffle is basically just shuffling the list of people
    random.shuffle(people)
    return people

# Created a drop-down menu, and filled it up with a random list of people from the API
selected_person = StringVar(root)
people = get_people()
selected_person.set(people[0]) # Set the default value to the first person in the list
person_menu = OptionMenu(root, selected_person, *people)
person_menu.pack()

# Created a button that gets a quote for the selected person
def get_quote():
    # Send a GET request to the Quotable API to retrieve a quote for the selected person
    response = requests.get(f'https://api.quotable.io/quotes?author={selected_person.get()}')
    quote = response.json()['results'][0]['content']
    # Display the quote in a text box
    quote_text.delete('1.0', END) # Deleting any previous text
    quote_text.insert(END, quote)

get_quote_button = Button(root, text="Get Quote", command=get_quote)
get_quote_button.pack()

# Made a button that resets the drop-down menu with a new random list of people from the API
def reset_menu():
    people = get_people()
    selected_person.set(people[0])
    menu = person_menu['menu']
    menu.delete(0, END)
    for person in people:
        menu.add_command(label=person, command=lambda person=person: selected_person.set(person))
    person_menu['menu'] = menu

reset_button = Button(root, text="Reset Menu", command=reset_menu)
reset_button.pack()

quote_text = Text(root) #created a text box to display the quote
quote_text.pack()

root.mainloop()
```

```python
#GUI 2
from tkinter import *
from tkinter import ttk
import requests


url = 'https://api.weather.gov/gridpoints/BOU/60,76/forecast'
response = requests.get(url)
forecast = response.json()['properties']['periods']

root = Tk()
root.title('7-Day Forecast')

# One widget we haven't learned in class is the treeview widget which I used here to display the forecast
tree = ttk.Treeview(root, columns=('period', 'temp', 'short_forecast'))
tree.heading('#0', text='Day')
tree.heading('period', text='Period')
tree.heading('temp', text='Temperature')
tree.heading('short_forecast', text='Forecast')

# Adding the forecast data to the treeview
for day in forecast:
    tree.insert('', END, text=day['name'], values=(day['number'], day['temperature'], day['shortForecast']))

# I could also change the style for the treeview
style = ttk.Style()
style.theme_use('default')
style.configure('Treeview', font=('Arial', 12))
style.configure('Treeview.Heading', font=('Arial', 12, 'bold'))

# Added scrollbars to the treeview
scrollbar = ttk.Scrollbar(root, orient='vertical', command=tree.yview)
scrollbar.pack(side='right', fill='y')
tree.configure(yscrollcommand=scrollbar.set)

# Packed the treeview widget
tree.pack(fill=BOTH, expand=True)

root.mainloop()

#Help from https://tkdocs.com/tutorial/tree.html

#"A treeview widget displays a hierarchy of items and allows users to browse through it. 
#One or more attributes of each item can be displayed as columns to the right of the tree. 
#It can be used to build user interfaces similar to the tree display you'd find in file managers 
#like the macOS Finder or Windows Explorer. As with most Tk widgets, it offers incredible flexibility 
#so it can be customized to suit a wide range of situations."
```
