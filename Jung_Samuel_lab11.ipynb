```python
from tkinter import *
 
class GreetingApp(Tk):
    def __init__(self):
        super().__init__()
        self.title('Greeting Application')
        self.geometry("300x300")
 
        self.name_var = StringVar()
        self.name_var.trace('w', self.create_greeting_message) 
        #trace is another method for string variables which 
        #allows you to notify if a value is change, read, or deleted.
 
        self.create_widgets()
     
    def create_widgets(self):
        self.description_label = Label(self, text="Enter your name:")
        self.description_label.grid(column=0, row=0)
 
        self.entry = Entry(self, textvariable=self.name_var)
        self.entry.grid(column=1, row=0)
        self.entry.focus()
 
        self.greeting_label = Label(self)
        self.greeting_label.grid(column=0, row=1, columnspan=2)
     
    def create_greeting_message(self, *args):
        name_entered = self.name_var.get()
 
        greeting_message = ""
        if name_entered != "":
            greeting_message = "Hello " + name_entered
         
        self.greeting_label['text'] = greeting_message
 
if __name__ == "__main__":
    app = GreetingApp()
    app.mainloop()
```

```python
#basic dropdown menu

from tkinter import *
import tkinter as tk
import random

root = Tk()
root.title("Dropdown practice")
root.geometry("400x400")

item_list = ["Yellow", "Blue", "Green", "Purple", "Violet", "Black",
                "Orange", "Gray", "Misty Rose", "Snow2", "goldrenrod1", "hotpink"]
    
def chosen_option():
    color = value.get()
    result_label = Label(root, text=f"You've chosen {color}")
    root.configure(background=color)
    result_label.pack()
    
def change_options(): #complicated way to reset the option menu
    value.set("") #reseting the menu default
    dropdown["menu"].delete(0, "end") #deleting the menu's contents
    item_list = ["Orange", "Gray", "Misty Rose", "Snow2", "goldenrod1", "hotpink"]
    for i in item_list: #below had to use a method tk._setit
        dropdown["menu"].add_command(label=i, command=tk._setit(dropdown, i))
        
value = StringVar()
value.set("Select a Color")

dropdown = OptionMenu(root, value, *item_list)
choose_button = Button(root, text="Click to choose", command=chosen_option)
change_button = Button(root, text="Click to change list", command=change_options) #new button

dropdown.pack()
choose_button.pack()
change_button.pack()
root.mainloop()
```
